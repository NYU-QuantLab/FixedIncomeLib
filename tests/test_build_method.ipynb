## Test Model Build Method

```python
### create yield curve build method
build_method_type = 'YIELD_CURVE'
content = {
    'TARGET' : 'SOFR-1B',
    'OVERNIGHT INDEX FUTURE' : 'SOFR-FUTURE-3M',
    'OVERNIGHT INDEX SWAP' : 'USD-SOFR-OIS'
}
yc_build_method = qfCreateBuildMethod(build_method_type, content)
display(yc_build_method.display())
```

```python
### serialize / de-serialize
path = 'yc_build_method.pickle'
qfWriteBuildMethodToFile(yc_build_method, path)
display(yc_build_method.display())
yc_build_method_back = qfReadBuildMethodFromFile(path)
# check
display(yc_build_method_back.display())
# house keeping
os.remove(path)
```

## Test Build Method Collection

```python
### create a data collection
bm_list = [yc_build_method]
bm_collection = qfCreateModelBuildMethodCollection(bm_list)
display(bm_collection.display())
```

```python
### serialization / de-serialization
# serialize
path = 'build_method_collection.pickle'
qfWriteBuildMethodCollectionToFile(bm_collection, path)
# deserialize
build_method_collecion_back = qfReadBuildMethodCollectionFromFile(path)
# check
display(build_method_collecion_back.display())
# house keeping
os.remove(path)
```

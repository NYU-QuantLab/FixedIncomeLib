## Test Index Fixings

```python
### list index fixings
index_name = 'USD-LIBOR-BBA-3M'
qfListIndexFixings(index_name.lower())
```

```python
### list index fixings for a specific start date and end date
index_name = 'USD-LIBOR-BBA-3M'
qfReloadIndexFixings() # load all back for testing below
qfListIndexFixings(
    index_name,
    start_date='2025-06-17',
    end_date='2025-06-18'
)
```

```python
### insert index fixings
qfInsertIndexFixing(index_name, '2026-06-19', 0.02736)
display(qfListIndexFixings(index_name))
# insert more
dates = ['2026-06-20', '2026-06-21']
fixings = [0.02, 0.02]
qfInsertIndexFixing(index_name, dates, fixings)
display(qfListIndexFixings(index_name))
# remove index fixings
qfRemoveIndexFixings(index_name, dates)
display(qfListIndexFixings(index_name))
```

```python
### remove all fixings
# remove all index fixings
qfRemoveIndexFixings('*')
qfListAllIndexFixings()
```

```python
### remove a specific fixing
name = 'USD-LIBOR-BBA-3M'
qfReloadIndexFixings() # load all back for testing below
qfRemoveIndexFixings(name)
qfListAllIndexFixings(name)
qfListAllIndexFixings('*')
```
